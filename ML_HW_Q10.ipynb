```python
# Cheney Zhang Homework #1 Question #10

import numpy as np
import matplotlib.pyplot as plt

epochs = 100
initial_loss = 10
loss_decay = 0.95
training_losses = [initial_loss * (loss_decay ** epoch) for epoch in range(epochs)]
validation_losses = [loss * np.random.uniform(0.8, 1.2) for loss in training_losses]  # Add some noise

plt.plot(training_losses, label='Training Loss')
plt.plot(validation_losses, label='Validation Loss', linestyle='--')
plt.title('Training & Validation Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
