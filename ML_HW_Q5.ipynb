```python
# Cheney Zhang Homework #1 Question #5

import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Setting the parameters
learning_rates = [0.02, 0.05, 0.1, 0.2]
n_epochs = 100
batch_size = len(y)

# Initialize weights and bias
w = np.random.randn(1, 1)
b = np.random.randn()

# Store the loss values for each learning rate
loss_history = {lr: [] for lr in learning_rates}

# Training loop for each learning rate
for lr in learning_rates:
    # Reset weights and bias
    w = np.random.randn(1, 1)
    b = np.random.randn()

    for epoch in range(n_epochs):
        # Calculate predictions
        predictions = X.dot(w) + b

        # Calculate the loss
        errors = predictions - y
        loss = (errors**2).mean()
        loss_history[lr].append(loss)

        # Calculate gradients
        w_gradient = 2 * X.T.dot(errors) / batch_size
        b_gradient = 2 * errors.mean()

        # Update weights and bias
        w -= lr * w_gradient
        b -= lr * b_gradient

# Plotting the loss for each learning rate
for lr in learning_rates:
    plt.plot(loss_history[lr], label=f'Leanring Rate={lr}')

plt.title('Loss over Epochs for Different Learning Rates')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
